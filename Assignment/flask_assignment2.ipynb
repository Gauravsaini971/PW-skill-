## Question 1

HTTP defines a set of request methods to indicate the desired action to be performed for a given resource. The two most common HTTP methods are GET and POST.

GET method is used to retrieve data from the server. It is used to request data from a specified resource. It is a safe method and should only be used for retrieving data. The data sent using GET method is visible in the URL.

POST method is used to submit an entity to the specified resource, often causing a change in state or side effects on the server. It is used to send data to the server for processing. The data sent using POST method is not visible in the URL.

In summary, GET method is used for retrieving data while POST method is used for submitting data.

## Question 2

In Flask, the request object is used to access incoming request data. The request object is an instance of the Request class which contains the data received from the client. It is used to retrieve data sent by the client such as form data, JSON data, query parameters, etc.

The request object is automatically created by Flask when a request is received and is available to all view functions during that request.

## Question 3

In Flask, the redirect() function is used to redirect the user to a particular URL with the status code. When this function is called, a response object is returned and the redirection happens to the target location with the status code.

The redirect() function is useful in Flask because it allows you to tell your app which URL it should go to when a certain action is performed by an end-user.

## Question 4

Templates in Flask are files that contain static data as well as placeholders for dynamic data. They are used to create HTML pages dynamically. The render_template() function is used to render templates. It takes the name of the template file as its first argument and any additional arguments that should be passed to the template as keyword arguments.

The render_template() function is useful because it allows you to separate the presentation logic from the application logic. This makes it easier to maintain your code and reuse templates across multiple views.

```python
app = Flask(__name__)

@app.route('/postman_data1',methods=['POST'])
def math_ops2():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        
        return jsonify(result)

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
