## Question 5

To build URLs in Flask, we use the url_for() function. The name of the function is accepted as its first argument and any number of keyword arguments. The unknown variable parts in a website are then appended to the URL as query parameters.

```python
from flask import Flask, url_for
```

```python
@app.route("/index")
def index():
    return 'index'

@app.route("/login")
def login():
    return "login"

@app.route("/user/<username>")
def profile(username):
    return f"{username}\'s profile"

with app.test_request_context():
    print(url_for("index"))
    print(url_for("login"))
    print(url_for("login",next="/"))
    print(url_for("profile",username = 'gaurav'))
```
