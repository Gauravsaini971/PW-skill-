## Question 2

A scatter plot is a type of plot that displays values for two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis. Scatter plots are used to observe relationships between variables and can be used to identify trends or patterns in data. In Matplotlib, scatter plots can be created using the scatter() method in the pyplot module.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,s = y*3)
plt.title("Scatter Plot")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, ax = plt.subplots(4,1)

ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
ax[3].plot(x,y4)

plt.show()
```

## Question 4

A bar plot is a type of plot that displays categorical data with rectangular bars. The height or length of each bar is proportional to the value it represents. Bar plots are used to compare values across categories and can be used to identify trends or patterns in data. In Matplotlib, bar plots can be created using the bar() method in the pyplot module.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.title("Companies V/S Profit")
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.show()
```

```python
plt.barh(company,profit,color = "b")
plt.show()
```

## Question 5

A box plot is a type of plot that displays the distribution of data based on five number summary (â€œminimumâ€, first quartile (Q1), median, third quartile (Q3), and â€œmaximumâ€). It is also known as a box-and-whisker plot. Box plots are used to identify outliers and to compare distributions of data across categories. In Matplotlib, box plots can be created using the boxplot() method in the pyplot module .

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
```
