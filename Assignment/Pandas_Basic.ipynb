## Question 4

* A Series is the one dimensional array which contain any kind of data. It is the type of list in pandas that can have labels as keys. A series is also the data structure for a single column of DataFrame.

* A DataFrame isthe collection of series that share same number of index. It is the 2 dimensional form of data structure with columns of potentially different types.In Dataframe each columns of data is represented as a pandas series. DataFrame columns can have names but series columns cannot have name.
For example :-

```python
l = [i for i in range(1,11)]
ser = pd.Series(l)
df_1 = pd.DataFrame(l)
```
