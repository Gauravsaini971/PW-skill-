## Question 9

```python
df_02 = pd.DataFrame({'Values': [i for i in range(10,60,10)]})
```

```python
df_02
```

```python
def get_stats(df):
  stats = {}
  for i in list(df.columns):
    if i == "Values":
      stats["Mean"] = df[i].mean()
      stats["Median"] = df[i].median()
      stats["Standard Deviation"] = df[i].std()

  return stats
```

```python
get_stats(df_02)
```

## Question 10

```python
df_03 = pd.DataFrame({'Sales': [i for i in range(10,101,10)],
                   'Date': pd.date_range('2023-01-01', periods=10, freq='D')})
df_03
```

```python
def get_moving_avg(df):
   df["Moving Average"] = df.rolling(window = 7,min_periods=1).mean()
   return df
```

```python
get_moving_avg(df_03)
```
